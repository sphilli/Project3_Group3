```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("world_country_and_usa_states_latitude_and_longitude_values.csv")
df1.nunique
df1.head(20)
```

```python
df1.nunique()
```

```python
# renamed columns for ease of merge
df1.rename(columns={"country":"Country"}, inplace=True)
df1.info()
```

```python
# read into csv
df2 = pd.read_csv("List of game studios in Europe and the CIS 2020.csv")
df2.info()
```

```python
# merged the CSVs on "Country" with a right join
merge_df = pd.merge(df1,df2,on="Country", how="right")
merge_df.head()
```

```python
merge_df.info()
```
